```python
import shutil
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
import random
import threading
```

```python
done = open("done.txt", "r")
done = done.read()
done = done.split("\n")
done = [int(item) for item in done[:-1]]

notfound = open("notfound.txt", "r")
notfound = notfound.read()
notfound = notfound.split("\n")
notfound = [int(item) for item in notfound[:-1]]
```

```python
ids = list(range(855388))
ids= list(set(ids)-set(done))
ids= list(set(ids)-set(notfound))
random.shuffle(ids)
len(ids)
```

```python
linksDone =list(range(855388))[::-1]
linksDone = linksDone[:-len(done)]
def scrapNews():
    for i in range(10):
        try:
            profile = webdriver.FirefoxProfile()
            profile.set_preference('permissions.default.image', 2)
            profile.set_preference("javascript.enabled", False);
            options = Options()
            options.add_argument("-headless")
            options.profile = profile
            service = Service(executable_path=shutil.which("geckodriver"))
            id = ids.pop()

            url = "https://news.am/arm/news/"+str(id)+".html"
            driver = webdriver.Firefox(service=service, options=options)
            try:
                driver.get(url)
            except:
                pass
            title = driver.find_element(By.CLASS_NAME, "article-title").text
            article = driver.find_elements(By.CLASS_NAME, "article-body")[1].text
            driver.close()

            f = open("news.txt","a")
            f.write(title+'\n')
            article = article.replace('\n\n','\n')
            f.write(article+'\n\n')
            f.close()

            f = open("done.txt","a")
            f.write(str(id)+'\n')
            f.close()
            
            print(i,id,linksDone.pop())
        except:
            f = open("notfound.txt","a")
            f.write(str(id)+'\n')
            f.close()

            driver.close()
            print(i,id,"not found")
            pass
```

```python
threading.Thread(target=scrapNews).start()
threading.Thread(target=scrapNews).start()
threading.Thread(target=scrapNews).start()
threading.Thread(target=scrapNews).start()
threading.Thread(target=scrapNews).start()
threading.Thread(target=scrapNews).start()
threading.Thread(target=scrapNews).start()
threading.Thread(target=scrapNews).start()
threading.Thread(target=scrapNews).start()
threading.Thread(target=scrapNews).start()
```
