```python
import re
from collections import Counter
```

```python
with open('../scraper/news.txt', 'r', encoding='utf-8',errors='ignore') as f:
    text = f.read()
text = text.replace('։',':',-1)
text = text.replace('․','.',-1)
text = text.replace('–','-',-1)

```

```python
chars = sorted(list(set(text)))
vocab_size = len(chars)
```

```python
chars,vocab_size
```

```python
nums = []
pattern = '|'.join(map(re.escape, chars))
matches = re.findall(pattern, text)
occurrences = Counter(matches)
```

```python
occurrences
```

```python
armenian="աբգդեզէըթժիլխծկհձղճմյնշոչպջռսվտրցւփքօֆև"
```

```python
armenian = armenian + armenian.upper()
symbols = "ՙ՚՛՜՝՞?! ,:-«».`)(+/%[]\n"
```

```python
nums = "0123456789"
eng = "abcdefghijklmnopqrstuvwxyz…\'\"´;@#)_—•"
eng = eng + eng.upper()
```

```python
rus = "абвгдеёжзийклмнопрстуфхцчшщъыьэюя"
rus = rus+rus.upper()

needed = armenian+nums+eng+rus+symbols
delete = set(chars) - set(needed)
```

```python
pattern2 = '|'.join(map(re.escape, delete))
matches2 = re.findall(pattern2, text)
occurrences2 = Counter(matches2)
```

```python
occurrences2
```

```python
len(text)
```

```python
a =''
for elem in delete:
    a+=elem
print(a)
```

```python
translation_table = str.maketrans('', '', a)
result_text = text.translate(translation_table)
```

```python
len(result_text)
```

```python
new_chars = sorted(list(set(result_text)))
new_vocab_size = len(new_chars)
```
